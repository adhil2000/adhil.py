```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
def getData(filename='input.txt'):
	return np.loadtxt(filename)
```

```python
data = getData('distA.csv')

stats.probplot(data, dist = 'norm', plot=plt)
plt.show()
```

The norm distribution is the best option for distA.csv. The probability plot of the data is very close to the theoretical distribution line unlike the other distribution options.

```python
data = getData('distB.csv')

stats.probplot(data, dist = 'laplace', plot=plt)
plt.show()
```

The laplace distribution is the best option for distB.csv. The probability plot of the data is very close to the theoretical distribution line unlike the other distribution options.
